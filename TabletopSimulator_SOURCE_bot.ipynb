```python
import cv2
from pathlib import Path
import pytesseract as pyt
pyt.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text_test = pyt.image_to_string('images/AntiFairy 1.jpg',
                                config = '--psm 3 tessedit_char_whitelist=01234546789')
print(text_test)
```

```python
test_image = cv2.imread('images/AntiFairy 1.jpg', 0)
img_thresh = cv2.threshold(test_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
data = pyt.image_to_string(img_thresh)#, config = '--psm 3')
print(data)
```
